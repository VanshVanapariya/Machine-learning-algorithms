```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
df = pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```python
x = df
y = data.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(np.array(y_train).reshape(-1,1))
```

```python
regressor = DecisionTreeRegressor()
```

```python
parameter={
 'criterion':['squared_error','friedman_mse','absolute_error','poisson'],  # friedman_mse : A variation of MSE used in gradient boosting. Slightly better for ensemble models.
                                                                           # poisson : Used for count data (non-negative integers), like predicting number of events.
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
  'max_features':['auto', 'sqrt', 'log2']
}
```

```python
tcv = GridSearchCV(regressor,parameter,cv=5,scoring='neg_mean_squared_error')
```

```python
tcv.fit(x_train,y_train)
```

```python
tcv.best_params_
```

```python
print(tcv.best_score_)
```

```python
y_pred = tcv.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
print("R²:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
