```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
X = df[['year']]
y = df[['per capita income (US$)']]

y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred_sc = regressor.predict(X_test)
y_pred = sc_y.inverse_transform(y_pred_sc)
```

```python
plt.scatter(X_test,y_test,color='red')
plt.plot(X_test,y_pred,color='lightgreen')
plt.scatter(X_test,y_pred,color='darkblue')

# plt.axvline(x=0, color='black', linestyle='--', linewidth=1.2, label='Origin')
# plt.axhline(y=19318, color='black', linewidth=1.2, label='intercept')

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('income prediction')
plt.legend()
plt.show()
```
